```python
import torch
import pandas as pd
from anomalib.engine import Engine
from anomalib.data import Folder
from anomalib.models import EfficientAd
import numpy as np
from sklearn.metrics import roc_auc_score

# DATA_ROOTì™€ OUTPUT_ROOT ì„¤ì •
DATA_ROOT = PROJECT_ROOT / "dataset" / "MMAD"
OUTPUT_ROOT = PROJECT_ROOT / "output" / "EfficientAd"

test_results_list = []
categories = [
    "cigarette_box", "drink_bottle", "drink_can", "food_bottle",
    "food_box", "food_package"
]

for category in categories:
    cat_path = DATA_ROOT / "GoodsAD" / category
    test_path = cat_path / "test"

    # 1. ë¶ˆëŸ‰ í´ë” ì´ë¦„ ë™ì  ì°¾ê¸°
    defect_types = [d.name for d in test_path.iterdir() if d.is_dir() and d.name != "good"]
    defect_name = defect_types[0] if defect_types else None

    if not defect_name:
        continue

    print(f"ğŸš€ [{category}] í…ŒìŠ¤íŠ¸ ì¤‘... (ë¶ˆëŸ‰ ìœ í˜•: {defect_name})")

    # 2. ëª¨ë¸ ìƒì„± ë° í•´ë‹¹ ì¹´í…Œê³ ë¦¬ ê°€ì¤‘ì¹˜ ë¡œë“œ
    model = EfficientAd()
    ckpt_path = OUTPUT_ROOT / category / "v0/weights/lightning/model.ckpt"

    if ckpt_path.exists():
        # state_dict ë¡œë“œ
        checkpoint = torch.load(ckpt_path, map_location="cpu")
        model.load_state_dict(checkpoint["state_dict"])
        print(f"âœ… ê°€ì¤‘ì¹˜ ë¡œë“œ ì™„ë£Œ: {ckpt_path.name}")
    else:
        print(f"âŒ ì²´í¬í¬ì¸íŠ¸ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŠµë‹ˆë‹¤: {ckpt_path}")
        continue

    # 3. ë°ì´í„° ëª¨ë“ˆ ìƒì„±
    datamodule = Folder(
        name=category,
        root=str(cat_path),
        normal_dir="train/good",
        abnormal_dir=f"test/{defect_name}",
        normal_test_dir="test/good",
        mask_dir=f"ground_truth/{defect_name}",
        extensions=[".jpg", ".jpeg", ".JPG", ".png", ".PNG"],
        eval_batch_size=32
    )
    datamodule.image_size = (256, 256)

    # 4. ì—”ì§„ ì„¤ì • (ì„ê³„ê°’ ì—ëŸ¬ ë°©ì§€ë¥¼ ìœ„í•´ ë‹¨ìˆœí•˜ê²Œ ì„¤ì •)
    engine = Engine(
        accelerator="auto",
        devices=1,
        logger=False,
        default_root_dir="temp_results",
    )

    # 5. í…ŒìŠ¤íŠ¸ ìˆ˜í–‰
    metrics = engine.test(
            model=model,
            datamodule=datamodule,
            ckpt_path=str(ckpt_path)
        )    # í…ŒìŠ¤íŠ¸ ê²°ê³¼ ì‹œê°í™” (ì´ë¯¸ì§€ í•œ ì¥ í™•ì¸ìš©)

    predictions = engine.predict(
            model=model,
            datamodule=datamodule,
            ckpt_path=str(ckpt_path)
        )

    # ìˆ˜ë™ AUROC ê³„ì‚°ì„ ìœ„í•œ ë¦¬ìŠ¤íŠ¸
    all_scores = []
    all_labels = []

    for batch in predictions:
        # 1. ì ìˆ˜ í‚¤ ì°¾ê¸°
        score_key = "pred_score" if "pred_score" in batch.keys() else "pred_scores"

        # 2. ë¼ë²¨(ì •ë‹µ) í‚¤ ì°¾ê¸° (gt_label í˜¹ì€ label í™•ì¸)
        # ë§Œì•½ ì•„ë˜ ë‘ ê°œê°€ ë‹¤ ì—†ë‹¤ë©´ print(batch.keys())ë¡œ í™•ì¸ì´ í•„ìš”í•©ë‹ˆë‹¤.
        if "gt_label" in batch.keys():
            label_key = "gt_label"
        elif "label" in batch.keys():
            label_key = "label"
        else:
            # ë§ˆì§€ë§‰ ìˆ˜ë‹¨: í‚¤ ëª©ë¡ì„ ì¶œë ¥í•˜ê³  ì—ëŸ¬ ë°©ì§€
            print(f"âš ï¸ ë¼ë²¨ í‚¤ë¥¼ ì°¾ì„ ìˆ˜ ì—†ìŒ. ì‚¬ìš© ê°€ëŠ¥í•œ í‚¤: {batch.keys()}")
            continue

        scores = batch[score_key].cpu().numpy()
        labels = batch[label_key].cpu().numpy()

        all_scores.extend(scores)
        all_labels.extend(labels)

    # ë¦¬ìŠ¤íŠ¸ê°€ ë¹„ì–´ìˆì§€ ì•Šì€ì§€ í™•ì¸ í›„ ê³„ì‚°
    if all_scores:
        all_scores = np.array(all_scores)
        all_labels = np.array(all_labels)

        # Min-Max ìŠ¤ì¼€ì¼ë§
        norm_scores = (all_scores - all_scores.min()) / (all_scores.max() - all_scores.min() + 1e-10)

        from sklearn.metrics import roc_auc_score
        manual_image_auroc = roc_auc_score(all_labels, norm_scores)
        print(f"ğŸš€ [{category}] ìˆ˜ë™ ì •ê·œí™” AUROC: {manual_image_auroc:.4f}")
    else:
        print(f"âŒ [{category}] ë°ì´í„°ë¥¼ ìˆ˜ì§‘í•˜ì§€ ëª»í–ˆìŠµë‹ˆë‹¤.")


    # 6. ê²°ê³¼ ì •ë¦¬
    m = metrics[0]
    image_auroc = m.get("image_AUROC", m.get("ImageAUROC", 0))
    pixel_auroc = m.get("pixel_AUROC", m.get("PixelAUROC", 0))

    test_results_list.append({
        "Category": category,
        "Image AUROC": round(float(image_auroc), 4),
        "Pixel AUROC": round(float(pixel_auroc), 4)
    })

# 7. ê²°ê³¼ ì¶œë ¥
print("\nğŸ“Š [GoodsAD] ìµœì¢… ì„±ëŠ¥ ì§€í‘œ")
df_final = pd.DataFrame(test_results_list)
print(df_final)
```

```python
import torch
from anomalib.models import EfficientAd
from anomalib.models.image.efficient_ad.torch_model import EfficientAdModelSize
from scripts.train_anomalib import Anomalibs
from scripts.VIS_RAG import VisualRAG

# 1. ìµœì‹  PyTorch ë³´ì•ˆ ì„¤ì • (í•œ ë²ˆë§Œ ì‹¤í–‰í•˜ë©´ ë¨)
torch.serialization.add_safe_globals([EfficientAdModelSize])

# 2. ì´ˆê¸° ì„¤ì •
app = Anomalibs(config_path="configs/runtime.yaml")
categories = app.get_all_categories()

# 3. ë£¨í”„ ì‹œì‘
for dataset, category in categories:
    print(f"\nğŸš€ [ì‘ì—… ì‹œì‘] {dataset} - {category}")

    # ì²´í¬í¬ì¸íŠ¸ ê²½ë¡œ í™•ì¸
    ckpt_path = app.get_ckpt_path(dataset, category)
    if not ckpt_path or not ckpt_path.exists():
        print(f"âš ï¸ {category}ì˜ ì²´í¬í¬ì¸íŠ¸ê°€ ì—†ì–´ ê±´ë„ˆëœë‹ˆë‹¤.")
        continue

    # ëª¨ë¸ ë¡œë“œ ë° RAG ê°ì²´ ì´ˆê¸°í™”
    model = EfficientAd.load_from_checkpoint(ckpt_path).to(app.device)
    rag = VisualRAG(model, app.device)

    # STEP 1: RAG DB êµ¬ì¶• ë° ì €ì¥ (ì´ë¯¸ ìˆë‹¤ë©´ load_dbë¡œ ëŒ€ì²´ ê°€ëŠ¥í•˜ì§€ë§Œ, ì²˜ìŒì—” êµ¬ì¶• í•„ìš”)
    dm_kwargs = app.get_datamodule_kwargs()
    datamodule = app.loader.get_datamodule(dataset, category, **dm_kwargs)
    rag.build_db(datamodule)
    rag.save_db(dataset, category)

    # STEP 2: ê¸°ì¡´ ì˜ˆì¸¡ ê²°ê³¼ì™€ ë§¤ì¹­í•˜ì—¬ í†µí•© JSON ìƒì„±
    # (ì´ë¯¸ app.predict_all()ì„ ëŒë ¤ì„œ predictions.jsonì´ ìˆë‹¤ëŠ” ì „ì œ í•˜ì— ì‘ë™í•©ë‹ˆë‹¤)
    rag.integrate_with_predictions(dataset, category)

    # ë©”ëª¨ë¦¬ ì •ë¦¬
    del model, rag
    torch.cuda.empty_cache()

print("\nâœ¨ ëª¨ë“  ì¹´í…Œê³ ë¦¬ì— ëŒ€í•œ RAG í†µí•© ì‘ì—…ì´ ì™„ë£Œë˜ì—ˆìŠµë‹ˆë‹¤!")
```
